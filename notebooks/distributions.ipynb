```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

* Plotting uniform distribution

```python
uniform_distri = np.random.uniform(1, 100, 10000)
```

```python
sns.histplot(uniform_distri)
```

* Plotting normal distribution

```python
normal_distri = np.random.normal(size=10000)
```

```python
sns.histplot(normal_distri)
plt.plot()
```
