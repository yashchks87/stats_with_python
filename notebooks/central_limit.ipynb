```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
uni_dist = np.random.uniform(1, 100, 10000)
```

```python
uni_dist
```

```python
plt.hist(uni_dist)
```

```python
sampled_ = []
for x in range(250):
    sampled_.append(np.random.choice(uni_dist, replace=False, size = 40))
```

```python
mean_vals = []
for x in range(250):
    mean_vals.append(np.mean(sampled_[x]))
```

```python
plt.hist(mean_vals)
plt.plot()
```
